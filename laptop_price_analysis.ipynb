```python
# Display the first 10 rows of the dataset
df.head(10)
```

```python
# Calculate and display average laptop price by TypeName
df.groupby('TypeName')['Price'].mean().sort_values(ascending=False)
```

```python
# Get the maximum laptop price for each company
df.groupby('Company')['Price'].max().sort_values(ascending=False)
```

```python
# Create a pivot table to show average prices by Type and Company
df.pivot_table(index='TypeName', columns='Company', values='Price', aggfunc='mean')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style="whitegrid")

# Bar plot: Average Price by Company
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Company', y='Price', estimator='mean', errorbar=None, palette='viridis')
plt.title('Average Laptop Price by Company')
plt.xticks(rotation=45)
plt.ylabel('Average Price')
plt.tight_layout()
plt.show()
```

```python
# Histogram: RAM distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Ram'], bins=5, kde=True, color='skyblue')
plt.title('Distribution of RAM')
plt.xlabel('RAM (GB)')
plt.show()

# Histogram: Weight distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Weight'], bins=6, kde=True, color='salmon')
plt.title('Distribution of Laptop Weight')
plt.xlabel('Weight (kg)')
plt.show()
```

```python
# Touchscreen vs Price
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Touchscreen', y='Price', palette='Set2')
plt.title('Price Comparison: Touchscreen vs Non-Touchscreen')
plt.xticks([0, 1], ['No Touchscreen', 'Touchscreen'])
plt.show()
```

```python
# Top 5 by price
top5 = df.sort_values(by='Price', ascending=False).head(5)
print("Top 5 Most Expensive Laptops:\n", top5[['Company', 'TypeName', 'Price']])
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Cpu brand', y='Price', palette='muted')
plt.xticks(rotation=45)
plt.title('Laptop Price by CPU Brand')
plt.show()
```
